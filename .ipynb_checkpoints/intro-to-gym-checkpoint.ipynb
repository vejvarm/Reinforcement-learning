# Introduction to OpenAI gym
Following instructions from <a href="https://gym.openai.com/docs/">documentation</a>

```python
import gym
import time
```

```python
env = gym.make('CartPole-v0')
for i_episode in range(20):
    observation = env.reset()
    for t in range(100):
        env.render()
        print(observation)
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        if done:
            print("Episode finished after {} timesteps".format(t+1))
            break
            
env.close()
```

```python
env = gym.make('CartPole-v0')
print(env.action_space)
#> Discrete(2) ==> either 0 or 1 (left or right)
print(env.observation_space)
#> Box(4,)
```

```python
print(env.observation_space.high)
print(env.observation_space.low)
env.close()
```

For CartPole-v0 one of the actions applies force to the left, and one of them applies force to the right. (Can you figure out which is which?)

```python
env = gym.make('CartPole-v0')

state = env.reset()
print(state)
for i in range(100):
    env.render()
    
    """ states:
    0 ... cart position
    1 ... cart velocity
    2 ... pole angle
    3 ... pole velocity
    """
    if state[2] < 0 and state[3] > -0.001:
        action = 1 # go right
    elif state[2] < 0 and state[3] > 0.001:
        action = 0 # go left
    elif state[2] > 0 and state[3] > -0.001:
        action = 1 # go right
    elif state[2] > 0 and state[3] > 0.001:
        action = 0 # go left
    elif state[2] < 0:
        action = 0
    elif state[2] > 0:
        action = 1
    
    state, reward, done, _ = env.step(action) # 0 is left, 1 is right
    
    print(state)
    
    time.sleep(0.1)
    
    if done:
        break

env.close()
```
